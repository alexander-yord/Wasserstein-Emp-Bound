```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dataclasses import dataclass

@dataclass
class PendulumParams:
    G: float = 9.8
    L1: float = 2.0
    L2: float = 2.0
    M1: float = 1.0
    M2: float = 1.0
    t_stop: float = 1000.0
    dt: float = 0.01

    @property
    def total_length(self) -> float:
        return self.L1 + self.L2

search_dir = "paths2"
search_pattern = os.path.join(search_dir, "*.csv")
csv_files = glob.glob(search_pattern)

# Indices of trajectories we want
traj_indices = [101, 378]

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharex=True, sharey=True)

for ax, idx in zip(axes, [101, 378]):
    target_file = csv_files[idx]
    df = pd.read_csv(target_file)

    w1 = df.iloc[:, 1].values
    w2 = df.iloc[:, 3].values

    ax.plot(w1, w2, marker='.', linestyle='None', markersize=1, alpha=0.4)
    ax.set_title(f"Trajectory {1 if idx == 101 else 2}")
    ax.set_xlabel(r'$\dot\theta_1$ (rad)')
    ax.set_ylabel(r'$\dot\theta_2$ (rad)')
    ax.grid(True)
    ax.set_aspect('equal')

plt.suptitle(r'Phase Plots: $\dot\theta_1$ vs $\dot\theta_2$', fontsize=14)
plt.subplots_adjust(wspace=-0.4)
plt.show()
```
