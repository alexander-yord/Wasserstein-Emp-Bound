# Gaussian Process Graphs

Load and plot `theta1_distributions.npz` for `paths2`.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np

npz_path = Path("gaussian_empirircal_process_output2/paths2/theta1_distributions.npz")
if not npz_path.exists():
    raise FileNotFoundError(f"Missing file: {npz_path}")

data = np.load(npz_path)
predicted = data["predicted_integrals"]
empirical = data["empirical_scaled_w"]

lo = min(empirical.min(), predicted.min())
hi = max(empirical.max(), predicted.max())
edges = np.linspace(lo, hi, 41)

fig, ax = plt.subplots(figsize=(10, 6), constrained_layout=True)
ax.hist(empirical, bins=edges, density=True, alpha=0.45, label="Empirical sqrt(q)W")
ax.hist(predicted, bins=edges, density=True, alpha=0.45, label="Predicted integral |G(t)| dt")
ax.set_title("theta1 distributions for paths2")
ax.set_xlabel("value")
ax.set_ylabel("density")
ax.legend(frameon=False)
plt.show()
```

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np

npz_path_1 = Path("gaussian_empirical_process_output1/paths1/w1_distributions.npz")
npz_path_2 = Path("gaussian_empirical_process_output2/paths1/w1_distributions.npz")

for path in (npz_path_1, npz_path_2):
    if not path.exists():
        raise FileNotFoundError(f"Missing file: {path}")

data_1 = np.load(npz_path_1)
data_2 = np.load(npz_path_2)

predicted_1 = data_1["predicted_integrals"]
empirical_1 = data_1["empirical_scaled_w"]
predicted_2 = data_2["predicted_integrals"]
empirical_2 = data_2["empirical_scaled_w"]

lo = min(empirical_1.min(), predicted_1.min(), empirical_2.min(), predicted_2.min())
hi = max(empirical_1.max(), predicted_1.max(), empirical_2.max(), predicted_2.max())
edges = np.linspace(lo, hi, 41)

fig, ax = plt.subplots(figsize=(11, 6), constrained_layout=True)
ax.hist(empirical_1, bins=edges, density=True, alpha=0.35, label="Output1 paths1 empirical")
ax.hist(predicted_1, bins=edges, density=True, alpha=0.35, label="Output1 paths1 predicted")
ax.hist(empirical_2, bins=edges, density=True, alpha=0.35, label="Output2 paths2 empirical")
ax.hist(predicted_2, bins=edges, density=True, alpha=0.35, label="Output2 paths2 predicted")
ax.set_title("theta1 distributions: output1/paths1 vs output2/paths2")
ax.set_xlabel("value")
ax.set_ylabel("density")
ax.legend(frameon=False)
plt.show()
```
